```python
for y in df.columns[df.columns!='Time']:
    plt.figure(figsize=(20,10))
    plt.plot(df['Time'], df[y], label=y)
    plt.legend()
    plt.show()
```

```python
for y in df.columns[df.columns!='Time']:
    plt.scatter(df['Time'], df[y], label=y)
    plt.legend()
    plt.show()
```

```python
plt.figure(figsize=(20, 10))
for y in df.columns[df.columns!='Time']:
    plt.plot(df['Time'], df[y], label=y)
plt.legend()
plt.show()
```

```python
ax, fig = plt.subplots(figsize=(20, 10))
for y in df.columns[df.columns!='Time']:
    if y == 'Engine RPM [RPM]':
        fig.plot(df['Time'], df[y]*0.085, label=y)
    else:
        fig.plot(df['Time'], df[y], label=y)
plt.legend()
plt.show()
```

```python
data = []
for y in df.columns[df.columns!='Time']:
    if y == 'Engine RPM [RPM]':
        data.append(go.Scatter(x=df['Time'], y=df[y]*0.085, mode='lines', name=y))
    else:
        data.append(go.Scatter(x=df['Time'], y=df[y], mode='lines', name=y))

pyo.plot(data, filename='data.html')
```
